```python
import pandas as pd
import re
```

```python
df=pd.read_csv(r'C:\Users\Zano\Downloads\Flask\train-data.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.drop('New_Price',axis=1,inplace=True)
```

```python
import re
re.search(r'[0-9]*.[0-9]*',df.Mileage[6018]).group()
```

```python
df.Mileage.dropna(inplace=True)
```

```python
df.Mileage=df.Mileage.apply(lambda x: re.search(r'[0-9]*.[0-9]*',x).group())
```

```python
df.Mileage.astype(float).quantile(0.01)
```

```python
df.drop(df[df.Mileage.astype(float)==0.0].index,axis=0,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.Engine=df.Engine.apply(lambda x: re.search(r'[0-9]*',x).group())
```

```python
df
```

```python
df.Engine.astype(int).quantile(0.95)
```

```python
df
```

```python
df.Power.isnull().sum()
```

```python
df.Power=df.Power.apply(lambda x: re.search(r'[0-9]*\.*[0-9]*',x).group())
```

```python
df
```

```python
df['Name'] = df['Name'].str.split(' ').str[0]
```

```python
df
```

```python
df.Name[df.Name=='ISUZU']='Isuzu'
```

```python
df.drop(df[df.Power==''].index,inplace=True)
```

```python
df.Kilometers_Driven=df.Kilometers_Driven.astype(int)
df.Mileage=df.Mileage.astype(float)
df.Engine=df.Engine.astype(int)
df.Power=df.Power.astype(float)
df.Price=df.Price.astype(float)
```

```python
df_train = pd.get_dummies(data=df, columns=['Name', 'Location','Fuel_Type','Transmission','Owner_Type'])
```

```python
y=df_train.Price
x=df_train.drop('Price',axis=1)
```

```python
l=[re.sub('Name_|Location_|Fuel_Type_|Owner_Type_|Transmission_','',i) for i in x.columns]
x.columns=l
```

```python
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
regressor_2 = RandomForestRegressor(random_state = 0)
regressor_2.fit(X_train, y_train)
```

```python
y_pred_2 = regressor_2.predict(X_test)
```

```python
regressor_2.score(X_test,y_test)
```
